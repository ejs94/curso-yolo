## **Como  a decomposio das camadas numa foto**

````python
pic = cv2.imread("../DATA/quafit.jpeg")
pic = cv2.cvtColor(pic, cv2.COLOR_BGR2RGB)
plt.imshow(pic)
````

 **Note que usamos a funo COLOR_BGR2RGB. Como vamos mostrar as imagens utilizando o matplotlib  necessrio inverter a ordem das camadas. Caso no faamos isso, as cores ficam distorcidas, vejam:**

````python
pic = cv2.imread("../DATA/quafit.jpeg")
plt.imshow(pic)
````
